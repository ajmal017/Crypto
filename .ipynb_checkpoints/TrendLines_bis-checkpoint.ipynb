## Numerical Differentiation method
However a far more elegant approach to this issue is to use the numerical derivative of the closing price to identify the points. The first derivative is the rate of change or effectively the momentum or velocity of the closing price, while the second derivative indicates the rate of change of the first derivative or its acceleration. Normal differentiation does not apply here as a discrete time series requires discrete numerical analysis tools. There are several advantages including that the numerical derivative will smooth out the data by considering all points in a given range from the point at which the rate of change is to be computed. For example the 5 points stencil method which considers some increment before and ahead, as well as a double increment before and ahead along with the point itself. The findiff library makes this computation easy and accurate even using higher order approximation methods:

```python
from findiff import FinDiff #pip install findiff
dx = 1 #1 day interval
d_dx = FinDiff(0, dx, 1)
d2_dx2 = FinDiff(0, dx, 2)
clarr = np.asarray(df_ema.close)
mom = d_dx(clarr)
momacc = d2_dx2(clarr)
```

```python
df_ema['Mom Acc'] = np.round(momacc,2)
df_ema['Mom'] = np.round(mom,2)
```

```python
h = df_ema.close.tolist()
```

```python
def get_extrema(isMin,mom, momacc, df):
    h = df.close.tolist()
    return [x for x in range(len(mom))
    if (momacc[x] > 0 if isMin else momacc[x] < 0) and
      (mom[x] == 0 or #slope is 0
        (x != len(mom) - 1 and #check next day
          (mom[x] > 0 and mom[x+1] < 0 and
           h[x] >= h[x+1] or
           mom[x] < 0 and mom[x+1] > 0 and
           h[x] <= h[x+1]) or
         x != 0 and #check prior day
          (mom[x-1] > 0 and mom[x] < 0 and
           h[x-1] < h[x] or
           mom[x-1] < 0 and mom[x] > 0 and
           h[x-1] > h[x])))]
```

```python
minimaIdxs, maximaIdxs = get_extrema(True, mom, momacc, df_ema), get_extrema(False, mom, momacc, df_ema)
```

```python
def get_bestfit3(x0, y0, x1, y1, x2, y2):
    xbar, ybar = (x0 + x1 + x2) / 3, (y0 + y1 + y2) / 3
    xb0, yb0, xb1, yb1, xb2, yb2 = x0-xbar, y0-ybar, x1-xbar, y1-ybar, x2-xbar, y2-ybar
    xs = xb0*xb0+xb1*xb1+xb2*xb2
    m = (xb0*yb0+xb1*yb1+xb2*yb2) / xs
    b = ybar - m * xbar
    ys0, ys1, ys2 = (y0 - (m * x0 + b)),(y1 - (m * x1 + b)),(y2 - (m * x2 + b))
    ys = ys0*ys0+ys1*ys1+ys2*ys2
    ser = np.sqrt(ys / xs)
    return m, b, ys, ser, ser * np.sqrt((x0*x0+x1*x1+x2*x2)/3)
```

```python
def get_bestfit(pts):
    xbar, ybar = [sum(x) / len (x) for x in zip(*pts)]
    def subcalc(x, y):
        tx, ty = x - xbar, y - ybar
        return tx * ty, tx * tx, x * x
    (xy, xs, xx) = [sum(q) for q in zip(*[subcalc(x, y) for x, y in pts])]
    m = xy / xs
    b = ybar - m * xbar
    ys = sum([np.square(y - (m * x + b)) for x, y in pts])
    ser = np.sqrt(ys / ((len(pts) - 2) * xs))
    return m, b, ys, ser, ser * np.sqrt(xx / len(pts))
```

## Naive method

```python
scale = (df_ema.close.max() - df_ema.close.min()) / len(df_ema)
errpct = 0.005 # 0.5%
fltpct=scale*errpct
```

```python
def get_trend(Idxs):
    trend = []
    for x in range(len(Idxs)):
        for y in range(x+1, len(Idxs)):
            for z in range(y+1, len(Idxs)):
                trend.append(([Idxs[x], Idxs[y], Idxs[z]],
                              get_bestfit3(Idxs[x], h[Idxs[x]],
                                           Idxs[y], h[Idxs[y]],
                                           Idxs[z], h[Idxs[z]])))
    return list(filter(lambda val: val[1][3] <= fltpct, trend))
```

```python
mintrend, maxtrend = get_trend(minimaIdxs), get_trend(maximaIdxs)
```

## Sorted Slope method

```python
def get_trend_opt(Idxs):
    slopes, trend = [], []
    
    for x in range(len(Idxs)): #O(n^2*log n) algorithm
        slopes.append([])
        for y in range(x+1, len(Idxs)):
            slope = (h[Idxs[x]] - h[Idxs[y]]) / (Idxs[x] - Idxs[y])
            slopes[x].append((slope, y))
            
    for x in range(len(Idxs)):
        slopes[x].sort(key=lambda val: val[0])
        CurIdxs = [Idxs[x]]
        for y in range(0, len(slopes[x])):
            CurIdxs.append(Idxs[slopes[x][y][1]])
            if len(CurIdxs) < 3: continue
            res = get_bestfit([(p, h[p]) for p in CurIdxs])
            if res[3] <= fltpct:
                CurIdxs.sort()
                if len(CurIdxs) == 3:
                    trend.append((CurIdxs, res))
                    CurIdxs = list(CurIdxs)
                else: CurIdxs, trend[-1] = list(CurIdxs), (CurIdxs, res)
            else: CurIdxs = [CurIdxs[0], CurIdxs[-1]] #restart search
    
    return trend
```

```python
mintrend2, maxtrend2 = get_trend_opt(minimaIdxs), get_trend_opt(maximaIdxs)
```

## Hough Line Transform method

```python
def make_image(Idxs):
    max_size = int(np.ceil(2/np.tan(np.pi / (360 * 5)))) #~1146
    m, tested_angles = df_ema.close.min(), np.linspace(-np.pi / 2, np.pi / 2, 360*5)
    height = int((df_ema.close.max() - m + 0.01) * 100)
    mx = min(max_size, height)
    scl = 100.0 * mx / height
    image = np.zeros((mx, len(df_ema))) #in rows, columns or y, x
    for x in Idxs:
        image[int((h[x] - m) * scl), x] = 255
    return image, tested_angles, scl, m
```

```python
def hough_points(pts, width, height, thetas):
    diag_len = int(np.ceil(np.sqrt(width * width + height * height)))
    rhos = np.linspace(-diag_len, diag_len, diag_len * 2)
    # Cache some resuable values
    cos_t = np.cos(thetas)
    sin_t = np.sin(thetas)
    num_thetas = len(thetas)
    # Hough accumulator array of theta vs rho
    accumulator =np.zeros((2 * diag_len, num_thetas), dtype=np.int64)
    # Vote in the hough accumulator
    for i in range(len(pts)):
        x, y = pts[i]
        for t_idx in range(num_thetas):
          # Calculate rho. diag_len is added for a positive index
            rho=int(round(x * cos_t[t_idx] + y * sin_t[t_idx])) + diag_len
            accumulator[rho, t_idx] += 1
    return accumulator, thetas, rhos
```

```python
def find_line_pts(Idxs, x0, y0, x1, y1):
    s = (y0 - y1) / (x0 - x1)
    i, dnm = y0 - s * x0, np.sqrt(1 + s*s)
    dist = [(np.abs(i+s*x-h[x])/dnm, x) for x in Idxs]
    dist.sort(key=lambda val: val[0])
    pts, res = [], None
    for x in range(len(dist)):
        pts.append((dist[x][1], h[dist[x][1]]))
        if len(pts) < 3: continue
        r = get_bestfit(pts)
        if r[3] > fltpct:
            pts = pts[:-1]
            break
        res = r
    pts = [x for x, _ in pts]
    pts.sort()
    return pts, res
```

```python
def houghpt(Idxs):
    max_size = int(np.ceil(2/np.tan(np.pi / (360 * 5)))) #~1146
    m, tested_angles = df_ema.close.min(), np.linspace(-np.pi / 2, np.pi / 2, 360*5)
    height = int((df_ema.close.max() - m + 1) * 100)
    mx = min(max_size, height)
    scl = 100.0 * mx / height
    acc, theta, d = hough_points([(x, int((h[x] - m) * scl)) for x in Idxs], mx, len(df_ema), np.linspace(-np.pi / 2, np.pi / 2, 360*5))
    origin, lines = np.array((0, len(df_ema))), []
    
    for x, y in np.argwhere(acc >= 3):
        dist, angle = d[x], theta[y]
        y0, y1 = (dist - origin * np.cos(angle)) / np.sin(angle)
        y0, y1 = y0 / scl + m, y1 / scl + m
        pts, res = find_line_pts(Idxs, 0, y0, len(df_ema), y1)
        if len(pts) >= 3: lines.append((pts, res))
    return lines
```

```python
mintrend3, maxtrend3 = houghpt(minimaIdxs), houghpt(maximaIdxs)
```

```python
from skimage.transform import hough_line, hough_line_peaks

def hough(Idxs): #pip install scikit-image
    image, tested_angles, scl, m = make_image(Idxs)
    h, theta, d = hough_line(image, theta=tested_angles)
    origin, lines = np.array((0, image.shape[1])), []
    for pts, angle, dist in zip(*hough_line_peaks(h, theta, d, threshold=2)):
        y0, y1 = (dist - origin * np.cos(angle)) / np.sin(angle)
        y0, y1 = y0 / scl + m, y1 / scl + m
        pts, res = find_line_pts(Idxs, 0, y0, image.shape[1], y1)
        if len(pts) >= 3: lines.append((pts, res))
    
    return lines
```

```python
mintrend4, maxtrend4 = hough(minimaIdxs), hough(maximaIdxs)
```

## Visual

```python
def get_trend_line(slope_support, intercept_support,
                   slope_resistance, intercept_resistance, df):
    
    x = np.array(range(len(df.index)))
    df['support'] = slope_support * x + intercept_support
    df['resistance'] = slope_resistance * x + intercept_resistance
    
    points = [(x[0],((df['resistance'].iloc[0]-df['support'].iloc[0])/2)+df['support'].iloc[0]),
              (x[-1],((df['resistance'].iloc[-1]-df['support'].iloc[-1])/2)+df['support'].iloc[-1])]
    x_coords, y_coords = zip(*points)
    A = np.vstack([x_coords,np.ones(len(x_coords))]).T
    m, c = np.linalg.lstsq(A, y_coords)[0]
    #print("Line Solution is y = {m}x + {c}".format(m=m,c=c))
    df['mid'] = m * x + c
    
    return df
```
