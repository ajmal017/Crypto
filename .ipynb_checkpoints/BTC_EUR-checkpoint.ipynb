```python
import krakenex
import datetime as datetime
import decimal
import time
import numpy as np
import pandas as pd
import talib
import warnings

warnings.filterwarnings("ignore")
```

```python
k = krakenex.API()
```

```python
pair = 'BTC/EUR'
df_trades = pd.DataFrame(columns = ['date', 'action', 'price'])


while True:
    print ('############# START ############')
    interval_ssl = 60
    bars_ssl = 100
    period_ssl = 10

    interval_ema = 5
    bars_ema = 200
    period_ema = 50
    
    tod = datetime.datetime.today()
    since_ssl = tod - datetime.timedelta(hours=0, minutes=interval_ssl * bars_ssl)
    since_ssl = since_ssl.timestamp()

    since_ema = tod - datetime.timedelta(hours = 0, minutes = interval_ema * bars_ema)
    since_ema = since_ema.timestamp()
    
    ret_ssl = k.query_public(method='OHLC', data = {'pair': pair, 'since': since_ssl, 'interval': interval_ssl})
    ret_ema = k.query_public(method='OHLC', data = {'pair': pair, 'since': since_ema, 'interval': interval_ema})

    array_ssl = ret_ssl['result'][list(ret_ssl['result'].keys())[0]]
    array_ema = ret_ema['result'][list(ret_ema['result'].keys())[0]]
    
    df_ssl = pd.DataFrame(data = array_ssl, columns = ['time','open','high','low','close','vwap','volume','count'])
    df_ema = pd.DataFrame(data = array_ema, columns = ['time','open','high','low','close','vwap','volume','count'])

    df_ssl = df_ssl.astype('float')
    df_ema = df_ema.astype('float')
    
    #function ssl
    low = df_ssl.low.values
    high = df_ssl.high.values

    df_ssl['sma_low'] = talib.SMA(low, timeperiod=period_ssl)
    df_ssl['sma_high'] = talib.SMA(high, timeperiod=period_ssl)

    df_ssl['time'] = df_ssl['time'].apply(lambda x:str(datetime.datetime.fromtimestamp(x)))

    df_ssl.set_index('time', inplace =True)
    df_ssl.dropna(inplace=True)

    df_ssl['green'] = pd.Series()
    df_ssl['red'] = pd.Series()

    for t in df_ssl.index:

        close = df_ssl.loc[t]['close']
        sma_high = df_ssl.loc[t]['sma_high']
        sma_low = df_ssl.loc[t]['sma_low']

        if close < sma_low:
            df_ssl.loc[slice(t,df_ssl.index[-1]),:]['red'] = df_ssl.loc[slice(t,df_ssl.index[-1]),:]['sma_high']
            df_ssl.loc[slice(t,df_ssl.index[-1]),:]['green'] = df_ssl.loc[slice(t,df_ssl.index[-1]),:]['sma_low']

        elif close > sma_high:
            df_ssl.loc[slice(t,df_ssl.index[-1]),:]['red'] = df_ssl.loc[slice(t,df_ssl.index[-1]),:]['sma_low']
            df_ssl.loc[slice(t,df_ssl.index[-1]),:]['green'] = df_ssl.loc[slice(t,df_ssl.index[-1]),:]['sma_high']

    # function ema 
    raw = df_ema.close.values
    ema = talib.EMA(raw, timeperiod=period_ema)

    df_ema['ema50'] = ema
    df_ema['ema50_to_close']  = (ema - raw)/raw 
    df_ema['time'] = df_ema['time'].apply(lambda x:datetime.datetime.fromtimestamp(x).strftime('%Y-%m-%d %H:%M:%S'))

    df_ema.set_index('time', inplace =True)
    df_ema.dropna(inplace=True)
            
    print (df_ssl[['close','green','red']].tail(1))
    print (df_ema[['close','ema50_to_close']].tail(1))
    

    if df_ssl.iloc[-1]['green'] > df_ssl.iloc[-1]['red']:
        print ('we are in ssl uptrend')
        print ('##############################################')
        
        #################################################################################
        #####################################BULLISH#####################################
        #################################################################################
        
        under_ema50_sec = (df_ema.loc[df_ema.index[-2],'ema50_to_close'] > 0)
        enter_ema50_actual = (df_ema.loc[df_ema.index[-1],'ema50_to_close'] < 0)
        engulfing_low = (df_ema.loc[df_ema.index[-2],'open'] - df_ema.loc[df_ema.index[-2],'close'])
        engulfing_high = (df_ema.loc[df_ema.index[-1],'close'] - df_ema.loc[df_ema.index[-1],'open'])
        hl = ((engulfing_high - engulfing_low)>0)
        
        green_candle_actual = (df_ema.loc[df_ema.index[-1],'open'] < df_ema.loc[df_ema.index[-1],'close'])
        red_candle_second = (df_ema.loc[df_ema.index[-2],'open'] > df_ema.loc[df_ema.index[-2],'close'])
            
        # ema conditons    
        if  under_ema50_sec and enter_ema50_actual and hl and green_candle_actual and red_candle_second:
                
            print ('we are in bullish engulfing pattern...its a buy')
            print ('price: {}'.format(df_ema.loc[df_ema.index[-1],'ema50_to_close'].values))
            
            trades = pd.read_csv('trades.csv')
            df_trades['date'] = [df_ema.index[-1]]
            df_trades['action'] = ['BUY']
            df_trades['price'] = [df_ema.loc[df_ema.index[-1],'ema50_to_close'].values]
            res = pd.concat([trades,df_trades])
            res.to_csv('trades.csv')
        else:
            
            print ('waiting for bullish engulfing pattern...')

        
        #################################################################################
        #################################################################################
        #################################################################################    
        
    if df_ssl.iloc[-1]['green'] < df_ssl.iloc[-1]['red']:
        print ('we are in ssl downtrend')

        #################################################################################
        #####################################BEARISH#####################################
        #################################################################################
        
        under_ema50_sec = (df_ema.loc[df_ema.index[-2],'ema50_to_close'] < 0)
        enter_ema50_actual = (df_ema.loc[df_ema.index[-1],'ema50_to_close'] > 0)
        engulfing_low = (df_ema.loc[df_ema.index[-2],'close'] - df_ema.loc[df_ema.index[-2],'open'])
        engulfing_high = (df_ema.loc[df_ema.index[-1],'open'] - df_ema.loc[df_ema.index[-1],'close'])
        hl = ((engulfing_high - engulfing_low)>0)
        
        green_candle_actual = (df_ema.loc[df_ema.index[-2],'open'] < df_ema.loc[df_ema.index[-2],'close'])
        red_candle_second = (df_ema.loc[df_ema.index[-1],'open'] > df_ema.loc[df_ema.index[-1],'close'])
            
        # ema conditons    
        if  under_ema50_sec and enter_ema50_actual and hl and green_candle_actual and red_candle_second:
                
            print ('we are in bearish engulfing pattern...its a sell')
            print ('price: {}'.format(df_ema.loc[df_ema.index[-1],'ema50_to_close'].values))

            trades = pd.read_csv('trades.csv')
            df_trades['date'] = [df_ema.index[-1]]
            df_trades['action'] = ['SELL']
            df_trades['price'] = [df_ema.loc[df_ema.index[-1],'ema50_to_close'].values]
            res = pd.concat([trades,df_trades])
            res.to_csv('trades.csv')
        
        else:
            
            print ('waiting for bearish engulfing pattern...')
        #################################################################################
        #################################################################################
        #################################################################################  
    
    
    next_time = array_ema[-1][0] + (interval_ema*60)
    sleep = float(next_time) - time.time()
    
    print (sleep/60, 'minutes sleep time')
    print ('############# END #############')
    
    time.sleep(sleep)
```
