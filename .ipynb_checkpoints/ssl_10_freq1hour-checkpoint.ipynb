```python
import krakenex
import datetime as datetime
import decimal
import time
import numpy as np
import pandas as pd
import talib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
import warnings
warnings.filterwarnings("ignore")
```

```python
pair = 'BTC/EUR'
# NOTE: for the (default) 1-minute granularity, the API seems to provide 720 points max
```

```python
stamp = datetime.datetime.today().timestamp()
tod = datetime.datetime.today()
print ('current timestamp: {}'.format(stamp))
print ('current decimal time: {}'.format(decimal.Decimal(time.time())))
print ('current date: {}'.format(tod))
```

```python
since = tod - datetime.timedelta(hours=720, minutes=0)
since
```

```python
k = krakenex.API()
k.load_key('kraken.key')
```

```python
ret = k.query_public(method='OHLC', data = {'pair': pair, 'since': since, 'interval': 60})
bars = ret['result'][list(ret['result'].keys())[0]]

df = pd.DataFrame(data = bars, columns = ['time','open','high','low','close','vwap','volume','count'])
df = df.astype('float')

period = 10
low = df.low.values
high = df.high.values

low_channel = talib.SMA(low, timeperiod=period)
high_channel = talib.SMA(high, timeperiod=period)

df['low_channel'] = low_channel
df['high_channel'] = high_channel

df['time'] = df['time'].apply(lambda x:str(datetime.datetime.fromtimestamp(x)))

df.set_index('time', inplace =True)
df.dropna(inplace=True)
```

```python
str(df.index[0])
```

```python
# Create figure and plot space
fig, ax = plt.subplots(figsize=(20, 10))
#ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M:%S'))
ax.xaxis.set_major_locator(mdates.DayLocator(interval=30)) 
# Add x-axis and y-axis
ax.plot(df.close,
        color='k')
ax.plot(df.low_channel,lw=1.5,
        color='red')

ax.plot(df.high_channel,lw=1.5,
        color='green')
# Set title and labels for axes
ax.set(xlabel="time",
       ylabel="Price",
       title=pair)

plt.setp(ax.get_xticklabels(), rotation=45)
plt.grid(True)
plt.show();
```
